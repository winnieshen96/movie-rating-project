```python
import pandas as pd
```

```python
filename = '../../Sample_Data/Raw/title.crew.tsv/data.tsv'
df = pd.read_csv(filename, sep='\t')
df.head()
```

```python
filename = '../../Sample_Data/For_Modeling/roi.plot.genresplitted.tsv'
movie_data = pd.read_csv(filename, sep='\t')
movie_data.head()
```

```python
movie_data_tconst_list = movie_data.tconst.values.tolist()
```

```python
df_tconst_list = df.tconst.values.tolist()
# num = 0
# for tconst in df_tconst_list:
#     if tconst in movie_data_tconst_list:
#          num += 1
# print(num)
```

```python
df = df.drop(df[~df['tconst'].isin(movie_data_tconst_list)].index)
len(df)
```

```python
df
```

```python
df['persons'] = df['directors']+','+df['writers']
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['persons'].values.tolist())
print(vectorizer.get_feature_names())
```

```python
print(X.toarray())
```

```python
def set_value(row_number, assigned_value): 
    if row_number in assigned_value:
        return assigned_value[row_number]
    else:
        return None
```

```python
roi_domestic_list = movie_data.loc[:, ['tconst', 'ROI_domestic']].values.tolist()
```

```python
roi_domestic_dict = {}
for i in roi_domestic_list:
    roi_domestic_dict[i[0]] = i[1]
```

```python
df['ROI_domestic'] = df['tconst'].apply(set_value, args =(roi_domestic_dict, ))
```

```python
df
```

```python
df.to_csv('../../Sample_Data/Processed/movie.crews.combined.tsv', sep='\t', index=False)
```

```python
import numpy as np
persons_lists = df.persons.values.tolist()
persons_lists = [persons.split(',') for persons in persons_lists]
v = persons_lists
l = [len(x) for x in v]
f, u = pd.factorize(np.concatenate(v))
n, m = len(v), u.size
i = np.arange(n).repeat(l)

dummies = pd.DataFrame(
    np.bincount(i * m + f, minlength=n * m).reshape(n, m),
    df.index, u
)

df = df.drop('persons', 1).join(dummies)
```

```python
df
```

```python
person_average_roi = {}
for person in vectorizer.get_feature_names():
    person_average_roi[person] = df[df[person]!=0]['ROI_domestic'].mean()
```
