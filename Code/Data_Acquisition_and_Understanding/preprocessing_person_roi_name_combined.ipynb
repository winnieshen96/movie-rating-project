```python
import pandas as pd
```

```python
crew_roi_data = pd.read_csv('../../Sample_Data/For_Modeling/person.averageroi.tsv', sep='\t')
```

```python
filename = '../../Sample_Data/Raw/name.basics.tsv/data.tsv'
person_data = pd.read_csv(filename, sep='\t')
person_data.head()
```

```python
joined_person_data = crew_roi_data.join(person_data.set_index('nconst'), how='left', on='personid')
joined_person_data.head()
```

```python
joined_person_data = joined_person_data.rename(columns={"personid": "person_id", "averageroi": "roi", "primaryName": "name"})
joined_person_data.head()
```

```python
joined_person_data.to_csv('../../Sample_Data/Processed/joined.person.roi.tsv', sep='\t', index=False)
```
