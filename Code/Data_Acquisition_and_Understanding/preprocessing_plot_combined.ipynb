```python
filename = '../../Sample_Data/Raw/story.tsv'
story = pd.read_csv(filename, sep='\t', index_col=False, 
                          names=['tconst', 'synopsis'])
story.head()
```

```python
len(story)
```

```python
story_list = story.values.tolist()
print(story_list[:5])
```

```python
story_dict = {}
for i in story_list:
    story_dict[i[0]] = i[1]
```

```python
def set_value(row_number, assigned_value): 
    return assigned_value[row_number]
```

```python
df_tconst_list = df.tconst.values.tolist()
story_tconst_list = story.tconst.values.tolist()
num = 0
for name in df_tconst_list:
    if name in story_tconst_list:
         num += 1
print(num)
```

```python
story.drop(story[~story['tconst'].isin(df_tconst_list)].index, inplace = True)
len(story)
```

```python
df['plot'] = df['tconst'].apply(set_value, args =(story_dict, ))
```
